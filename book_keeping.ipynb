```python
# Comparing fundings of the all companies:- 
```

```python
companies = ["HelloBooks", "Zoho Books", "QuickBooks", "HostBooks", "Vyapar"]
funding = [0, 0, 0, 8, 35]  # Funding in USD Millions
# Plot bar chart
plt.figure(figsize=(6,5))
bars = plt.bar(companies, funding, color="lightgreen")
# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, str(yval), ha='center', va='bottom')

plt.title("Funding of Bookkeeping Companies (USD M)")
plt.xlabel("Company")
plt.ylabel("Funding (in Millions USD)")
plt.show()
print("Point to notice:-")
print("*Zoho has never taken external funding, unique among SaaS companies of this scale\n*No public funding disclosed - early-stage startup likely bootstrapped\n*QuickBooks is backed by Intuit, a multibillion-dollar public company; traditional VC-style funding not applicable.")
```

```python
# Comparing the Employees working on the particular companies:-
```

```python
companies = ["HelloBooks", "Zoho Books", "QuickBooks", "HostBooks", "Vyapar"]
employees = [50, 15000, 10000, 200, 1000]  # Approx values from the dataset
colors=['#000000','#00FFFF','#FFFF00','#FF00FF','#FF0000']
# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(
    employees,
    colors=colors,
    labels=companies,
    autopct='%1.1f%%',
    startangle=140,
)

plt.title("Employees Distribution in Bookkeeping Companies")
plt.show()
```
